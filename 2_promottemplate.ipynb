### prompt template

```python
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI

# إنشاء القالب
prompt_template = PromptTemplate.from_template("what is the capital of {country}?")

# إنشاء prompt باستخدام القالب (هذه الخطوة كانت ناقصة)
prompt = prompt_template.format(country="Egypt")  # أو أي دولة تريد

# إنشاء الـ llm
llm = ChatGoogleGenerativeAI(model="gemini-3-flash-preview", temperature=0)

# استخدام prompt الصحيح
response = llm.invoke(prompt)
print(response.content)
```

### ChatPrompttemplate

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI

# 1. إنشاء قالب المحادثة
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that tells jokes."),
    ("user", "Tell me a joke about {topic}")
])

# 2. إنشاء النموذج
llm = ChatGoogleGenerativeAI(
    model="gemini-3-flash-preview", 
    temperature=0.7
)

# 3. تنسيق القالب مع الموضوع المطلوب
formatted_messages = prompt_template.format_messages(topic="cats")

# 4. إرسال الرسائل إلى النموذج
response = llm.invoke(formatted_messages)

# 5. طباعة الرد
print(response.content)
```

### MessagesPlaceHolder

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage, AIMessage

# 1. إنشاء قالب المحادثة مع MessagesPlaceholder
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that tells jokes."),
    MessagesPlaceholder("conversation_history"),  # مكان للرسائل السابقة
    ("user", "Tell me a joke about {topic}")
])

# 2. إنشاء النموذج
llm = ChatGoogleGenerativeAI(
    model="gemini-3-flash-preview", 
    temperature=0.7
)

# 3. إنشاء قائمة من الرسائل (تاريخ المحادثة)
conversation_messages = [
    HumanMessage(content="I like funny jokes!"),
    AIMessage(content="Great! I'll try to be extra funny then."),
    HumanMessage(content="Especially about animals")
]

# 4. تنسيق القالب مع المتغيرات
formatted_messages = prompt_template.format_messages(
    conversation_history=conversation_messages,  # قائمة الرسائل
    topic="cats"  # الموضوع
)

# 5. إرسال الرسائل إلى النموذج
response = llm.invoke(formatted_messages)

# 6. طباعة الرد
print("Response:")
print(response.content)
```
