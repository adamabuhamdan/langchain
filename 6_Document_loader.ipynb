```python
from langchain_community.document_loaders import JSONLoader
from langchain_unstructured import UnstructuredLoader  
from pathlib import Path
import os


files = Path("model").glob("*")

all_docs = []

for file in files:
    try:
        if file.suffix == ".json":
            un_loader = JSONLoader(str(file), jq_schema=".", text_content=False)
        else:
            # استخدام النسخة الجديدة من UnstructuredLoader
            un_loader = UnstructuredLoader(str(file))
        
        all_docs.extend(un_loader.load())
        print(f"✅ تم تحميل: {file.name}")
        
    except Exception as e:
        print(f"❌ خطأ في تحميل {file.name}: {e}")

print(f"\n📊 إجمالي المستندات: {len(all_docs)}")

if all_docs:
    print(f"\n📄 محتوى أول مستند:")
    print("-" * 50)
    print(all_docs[0].page_content[:500])  
    print("-" * 50)
```
