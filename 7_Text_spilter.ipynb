Character Text Spliter

```python
from langchain_text_splitters import CharacterTextSplitter

text = "Langchain simplifies building AI applications. It offers tools for splitting text and interacting with LLMs."

ch_splitter = CharacterTextSplitter(
    separator="", 
    chunk_size=20, 
    chunk_overlap=3
)

chunks = ch_splitter.split_text(text)

print(f"عدد الأجزاء: {len(chunks)}")
print(f"نوع النتيجة: {type(chunks)}")

for i, chunk in enumerate(chunks, 1):
    print(f"\nالجزء {i}: '{chunk}' (طول {len(chunk)} حرف)")
```

Recursive Character Text Spliter

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

path = "system_workflow.pdf"
pdf_loader = PyPDFLoader(path)
pdf_docs = pdf_loader.load()

rec_spliter = RecursiveCharacterTextSplitter(
  separators=["\n\n", "\n", ".", " ",""], chunk_size = 500, chunk_overlap= 30
)

pdf_chunk = rec_spliter.split_documents(pdf_docs)

print(len(pdf_chunk))
print("="*20)
print(pdf_chunk[10])
print("="*20)
print(pdf_chunk[10].page_content)
print("="*20)
```

MarkdownHeader Text Spliter

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import MarkdownHeaderTextSplitter

path = "text.md"

# تحميل الملف
md_loader = TextLoader(path, encoding='utf-8')
md_docs = md_loader.load()
md_text = md_docs[0].page_content

# تقسيم الملف بناءً على العناوين
md_splitter = MarkdownHeaderTextSplitter(
    headers_to_split_on=[
        ("#", "Title"),        # H1: العنوان الرئيسي
        ("##", "Section"),      # H2: الأقسام الرئيسية
        ("###", "Subsection")   # H3: الأقسام الفرعية
    ]
)

md_chunks = md_splitter.split_text(md_text)

print(f"📊 عدد الأجزاء: {len(md_chunks)}")
print("=" * 60)

for i, chunk in enumerate(md_chunks):
    print(f"\n📄 الجزء {i+1}:")
    print(f"   🏷️ الميتاداتا: {chunk.metadata}")
    print(f"   📝 المحتوى:")
    print("-" * 40)
    print(chunk.page_content.strip())
    print("-" * 40)
```
