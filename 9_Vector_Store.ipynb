Adding Documents

```python
from langchain_community.vectorstores import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_core.documents import Document
from dotenv import load_dotenv
import os

load_dotenv()

embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")

PERSIST_DIR = "Chroma_db"

# ✅ لا تحذف المجلد، فقط استخدمه إذا كان موجوداً
print(f"📁 استخدام المجلد: {PERSIST_DIR}")

# إنشاء قاعدة البيانات (ستضاف إلى المجلد الموجود أو تنشئ جديداً)
docs = [
    Document(page_content="Hello World", metadata={"source": "doc1", "type": "greeting"}),
    Document(page_content="World Economy", metadata={"source": "doc2", "type": "economics"})
]

ch_doc_db = Chroma.from_documents(
    documents=docs, 
    embedding=embeddings, 
    collection_name="My_docs", 
    persist_directory=PERSIST_DIR
)
# عرض المستندات بعد الإضافة الأولى
print("\n📄 المستندات بعد الإضافة الأولى:")
print("-" * 40)
all_docs = ch_doc_db.get()
documents = all_docs['documents']
metadatas = all_docs['metadatas']

for i, (doc, meta) in enumerate(zip(documents, metadatas), 1):
    print(f"{i}. {doc} (المصدر: {meta.get('source', 'غير معروف')})")
print(f"\n📊 عدد المستندات: {len(documents)}")

# المستندات الثانية
docs2 = [
    Document(page_content="Data Science", metadata={"source": "doc3", "type": "technology"}),
    Document(page_content="Machine Learning", metadata={"source": "doc4", "type": "technology"})
]

print("\n➕ إضافة مستندات جديدة:")
for doc in docs2:
    print(f"   • {doc.page_content} (المصدر: {doc.metadata['source']})")

# إضافة المستندات
ch_doc_db.add_documents(documents=docs2)

# عرض المستندات بعد الإضافة الثانية
print("\n📄 المستندات بعد الإضافة الثانية:")
print("-" * 40)
all_docs_updated = ch_doc_db.get()
documents_updated = all_docs_updated['documents']
metadatas_updated = all_docs_updated['metadatas']

for i, (doc, meta) in enumerate(zip(documents_updated, metadatas_updated), 1):
    print(f"{i}. {doc} (المصدر: {meta.get('source', 'غير معروف')})")
print(f"\n📊 إجمالي المستندات: {len(documents_updated)}")

# البحث الدلالي
print("\n" + "=" * 60)
print("🔍 البحث الدلالي")
print("=" * 60)

query = "World"
results = ch_doc_db.similarity_search(query, k=4)

print(f"\n🔎 البحث عن: '{query}'")
print("-" * 40)
for i, doc in enumerate(results, 1):
    print(f"{i}. {doc.page_content}")
    print(f"   المصدر: {doc.metadata.get('source', 'غير معروف')}")
    print(f"   النوع: {doc.metadata.get('type', 'غير معروف')}")
    print()

print("\n✅ تمت العملية بنجاح!")
```

## FAISS

Adding Text

```python
from langchain_community.vectorstores import FAISS 
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_core.documents import Document
from dotenv import load_dotenv
import os

load_dotenv()

# تهيئة embeddings
embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")

print("=" * 60)
print("📚 تجربة FAISS Vector Store")
print("=" * 60)

# 1. إنشاء قاعدة البيانات الأولى
print("\n📝 قاعدة البيانات الأولى:")
text_docs1 = ["Hello World 1", "World Economy 1"]
print(f"   النصوص: {text_docs1}")

faiss_db1 = FAISS.from_texts(
    texts=text_docs1, 
    embedding=embeddings
)

print("\n📄 المستندات في قاعدة البيانات الأولى:")
print("-" * 40)
# طريقة آمنة للوصول إلى المستندات
if hasattr(faiss_db1, 'docstore') and hasattr(faiss_db1.docstore, '_dict'):
    docs1 = list(faiss_db1.docstore._dict.values())
    for i, doc in enumerate(docs1, 1):
        print(f"{i}. {doc.page_content}")
        if hasattr(doc, 'metadata'):
            print(f"   الميتاداتا: {doc.metadata}")

# 2. إنشاء قاعدة البيانات الثانية
print("\n" + "=" * 60)
print("\n📝 قاعدة البيانات الثانية:")
text_docs2 = ["Data Science 2", "Machine Learning 2"]
print(f"   النصوص: {text_docs2}")

faiss_db2 = FAISS.from_texts(
    texts=text_docs2, 
    embedding=embeddings
)

print("\n📄 المستندات في قاعدة البيانات الثانية:")
print("-" * 40)
if hasattr(faiss_db2, 'docstore') and hasattr(faiss_db2.docstore, '_dict'):
    docs2 = list(faiss_db2.docstore._dict.values())
    for i, doc in enumerate(docs2, 1):
        print(f"{i}. {doc.page_content}")

# 3. دمج قاعدة البيانات الأولى في الثانية
print("\n" + "=" * 60)
print("\n🔄 دمج قاعدة البيانات الأولى في الثانية...")
faiss_db2.merge_from(faiss_db1)

print("\n📄 المستندات بعد الدمج:")
print("-" * 40)
if hasattr(faiss_db2, 'docstore') and hasattr(faiss_db2.docstore, '_dict'):
    docs_merged = list(faiss_db2.docstore._dict.values())
    for i, doc in enumerate(docs_merged, 1):
        print(f"{i}. {doc.page_content}")
print(f"\n📊 إجمالي المستندات بعد الدمج: {len(docs_merged)}")

# 4. اختبار البحث
print("\n" + "=" * 60)
print("🔍 اختبار البحث")
print("=" * 60)

search_queries = ["World", "Science", "Data", "Hello"]

for query in search_queries:
    print(f"\n🔎 البحث عن: '{query}'")
    print("-" * 40)
    
    results = faiss_db2.similarity_search(query, k=3)
    
    if results:
        for i, doc in enumerate(results, 1):
            print(f"{i}. {doc.page_content}")
            if hasattr(doc, 'metadata') and doc.metadata:
                print(f"   الميتاداتا: {doc.metadata}")
    else:
        print("   ❌ لا توجد نتائج")

# 5. حفظ واسترجاع FAISS (اختياري)
print("\n" + "=" * 60)
print("💾 حفظ قاعدة البيانات")
print("=" * 60)

# حفظ
save_path = "faiss_index"
faiss_db2.save_local(save_path)
print(f"✅ تم حفظ قاعدة البيانات في: {save_path}/")

# استرجاع
loaded_db = FAISS.load_local(save_path, embeddings, allow_dangerous_deserialization=True)
print(f"✅ تم استرجاع قاعدة البيانات")
print(f"   عدد المستندات بعد الاسترجاع: {len(loaded_db.docstore._dict)}")

print("\n" + "=" * 60)
print("✅ تمت العملية بنجاح!")
print("=" * 60)
```
